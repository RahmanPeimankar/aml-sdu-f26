## Main Ideas of Backpropagation

1. Using the **Chain Rule** to calculate derivatives.
2. Plugging the derivatives into **Gradient Descent** to optimize parameters.

In the next following slides, we will talk about how the **Chain Rule** and **Gradient Descent** apply to multiple parameters simultaneously, and introduce some fancy notations!

<center> 
<img src="img/Qimage-27.JPG" width="800"/>
    
**NOTE**: **Back propagation** start with the last parameters (**$b_3$**) and works backwards to estimate all of the other parameters.

To start from the back, let's assume that we already have <font color='green'>optimal values<font color='black'> for all of the parameters except for the last **Bias** term (**$b_3$**).
 
<center> 
<img src="img/Qimage-28.JPG" width="1000"/>

<center> 
<img src="img/Qimage-29.JPG" width="800"/>

<center>

* We don't know yet the optimal value for **$b_3$**, we have to give it an initial value. 
* Because **Bias** terms are usually initialized to **0**, we will set **$b_3=0$**.

We can quantify how good the <font color='green'>green curve<font color='black'> fits the data calculating the **Sum of the Squared Residulas**.
    
<center> 
<img src="img/Qimage-31.JPG" width="200"/> 
    
$\text{SSR} = (0 - (-2.6))^2 + (1 - (-1.61))^2 + (0 - (-2.61))^2 = 20.4$

* When **$b_3 = 0$**, the **SSR = 20.4**.
* If we increase **$b_3$** to **1**, then we would add **1** to the y-axis coordinates on the <font color='green'>**green curve** <font color='black'> and shift it up **1**.

$\text{SSR} = (0 - (-1.61))^2 + (1 - (-0.61))^2 + (0 - (-1.61))^2 = 7.8$

<center> 
<img src="img/Qimage-32.JPG" width="200"/>

If we increase $b_3$ again to **2** and **3**, we will get lower **SSR**.

$\text{Predicted}_i$ = <font color='green'> green curve <font color='black'>= <font color='blue'> blue <font color='black'>+ <font color='orange'> orange curve <font color='black'>+ $b_3$

We want to use **Gradient Descent** to optimize $b_3$ and that means we need to take the derivative of the **SSR** with respect to $b_3$ ($\frac{d\,SSR}{d\,b_3}$).

<center>
<img src="img/Qimage-39.JPG" width="200"/>



* And because the **SSR** are linked to $b_3$ by the **Predicted** values, we can use the **Chain Rule**!

$\text{SSR} = \sum_{i=1}^{n=3}(\text{Observed}_i - \text{Predicted}_i)^2$

$\text{Predicted}_i$ = <font color='green'> green curve <font color='black'>= <font color='blue'> blue <font color='black'>+ <font color='orange'> orange curve <font color='black'>+ $b_3$

$\frac{d\,SSR}{d\,b_3} = \frac{d\,SSR}{d\,Predicted} \times \frac{d\,Predicted}{d\,b_3}$

1) $\text{SSR} = \sum_{i=1}^{n=3}(\text{Observed}_i - \text{Predicted}_i)^2$

2) $\text{Predicted}_i$ = <font color='green'> green curve <font color='black'>= <font color='blue'> blue <font color='black'>+ <font color='orange'> orange curve <font color='black'>+ $b_3$
    
3) $\frac{d\,SSR}{d\,b_3} = \frac{d\,SSR}{d\,Predicted} \times \frac{d\,Predicted}{d\,b_3}$

$\frac{d\,SSR}{d\,Predicted} = \frac{d\,}{d\,Predicted} \sum_{i=1}^{n=3}(\text{Observed}_i - \text{Predicted}_i)^2 \\ \quad\quad\quad=\sum_{i=1}^{n=3}2\times(\text{Observed}_i - \text{Predicted}_i) \times \frac{d\,}{d\,Predicted}(\text{Observed - Predicted}) \\  \quad\quad\quad=\sum_{i=1}^{n=3}2\times(\text{Observed}_i - \text{Predicted}_i) \times -1 \\  \quad\quad\quad=\sum_{i=1}^{n=3}-2\times(\text{Observed}_i - \text{Predicted}_i)$

$\frac{d\,Predicted}{d\,b_3} = \frac{d\,}{d\,b_3} (\text{green curve}) = \frac{d\,}{d\,b_3} (\text{blue + orange} + b_3) = \frac{d\,blue}{d\,b_3} + \frac{d\,orange}{d\,b_3} + \frac{d\,b_3}{d\,b_3} = 0 + 0 + 1 = 1$

$\frac{d\,SSR}{d\,b_3} = \sum_{i=1}^{n=3}-2\times(\text{Observed}_i - \text{Predicted}_i) \times 1$

* So, we have the derivative of the **SSR** with respect to $b_3$.

$\frac{d\,SSR}{d\,b_3} = \frac{d\,SSR}{d\,Predicted} \times \frac{d\,Predicted}{d\,b_3} = \sum_{i=1}^{n=3}-2\times(\text{Observed}_i - \text{Predicted}_i) \times 1$


* This means we can plug this derivative into **Gradient Descent** to find the optimal value for $b_3$.

$\frac{d\,SSR}{d\,b_3} = -2 \times (\text{Observed}_1 - \text{Predicted}_1) \times 1\\ \quad\quad\quad + -2 \times (\text{Observed}_2 - \text{Predicted}_2) \times 1\\\quad\quad\quad + -2 \times (\text{Observed}_3 - \text{Predicted}_3) \times 1$ 

$\frac{d\,SSR}{d\,b_3} =-2 \times (0 - \text{Predicted}_1) \times 1\\ \quad\quad\quad + -2 \times (1 - \text{Predicted}_2) \times 1\\\quad\quad\quad + -2 \times (0 - \text{Predicted}_3) \times 1$

* How about the **Predicted** values?

* The values **Predicted** are given by the <font color='green'>**green curve** <font color='black'>.
* And, we get the **predicted** values on the <font color='green'>**green curve** <font color='black'>by running the **Dosages** through the **Neural Network**.

In conclusion, **Main Idea** for **Backpropagation** is that when a parameter like $b_3$ is unknown:

* We use the **Chain Rule** to calculate the derivative of the **Sum of the Squared Residulas (SSR)**

    * with respect to the unknown parameters (in this case $b_3$).

* Then we initialize the unknown parameters with a number (in this case we set $b_3 = 0$)

* And used **Gradient Descent** to optimize the unknown parameter

# Neural Nets in Practice

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(precision=3, suppress=True)
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=2)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
xlim = plt.xlim()
ylim = plt.ylim()
```

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train, y_train)
print(mlp.score(X_train, y_train))
print(mlp.score(X_test, y_test))
```

```python
xs = np.linspace(xlim[0], xlim[1], 1000)
ys = np.linspace(ylim[0], ylim[1], 1000)
xx, yy = np.meshgrid(xs, ys)
X_grid = np.c_[xx.ravel(), yy.ravel()]
plt.contour(xx, yy, mlp.predict_proba(X_grid)[:, 1].reshape(xx.shape), levels=[.5])
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

plt.xlim(xlim)
plt.ylim(ylim)
```

# Random State

```python
import warnings
warnings.filterwarnings(action='ignore')
                        
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
for ax, i in zip(axes.ravel(), range(10)):
    mlp = MLPClassifier(solver='lbfgs', random_state=i).fit(X_train, y_train)

    ax.contour(xx, yy, mlp.predict_proba(X_grid)[:, 1].reshape(xx.shape), levels=[.5])
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xticks(())
    ax.set_yticks(())
```

```python
# from keras.wrappers.scikit_learn import KerasClassifier
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

def make_model(optimizer="adam", hidden_size=(32,32)):
    model = Sequential([
        Dense(hidden_size[0], input_shape=(X_train.shape[1],), name='dense_1'),
        Activation('relu', name='activation_1'),
        Dense(hidden_size[1], name='dense_2'),
        Activation('relu', name='activation_2'),
#         Dropout(0.5)
        Dense(1, name='dense_3'),
        Activation('sigmoid', name='activation_3')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

clf = KerasClassifier(make_model)
param_grid = {'epochs': [10, 50, 100],  
              'model__hidden_size': [(4,4), (4,16), (16,16)]}
grid = GridSearchCV(clf, param_grid=param_grid, cv=5, return_train_score=True)
grid.fit(X_train, y_train)
```

```python
res = pd.DataFrame(grid.cv_results_)
res.pivot_table(index=["param_epochs", "param_model__hidden_size"],
                values=['mean_train_score', "mean_test_score"])
```
